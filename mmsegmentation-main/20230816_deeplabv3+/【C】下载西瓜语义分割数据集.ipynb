```python
# # 删除原有的数据集目录（如有）
# !rm -rf Watermelon87_Semantic_Seg_Mask.zip Watermelon87_Semantic_Seg_Mask
# 
# # 下载数据集压缩包
# !wget https://zihao-openmmlab.obs.cn-east-3.myhuaweicloud.com/20230130-mmseg/dataset/watermelon/Watermelon87_Semantic_Seg_Mask.zip
# 
# # 解压
# !unzip Watermelon87_Semantic_Seg_Mask.zip >> /dev/null
# 
# # 删除压缩包
# !rm -rf Watermelon87_Semantic_Seg_Mask.zip

# 删除原有的数据集目录（如有）
!rmdir /s /q Watermelon87_Semantic_Seg_Mask.zip
!rmdir /s /q Watermelon87_Semantic_Seg_Mask

# 下载数据集压缩包
!curl -o Watermelon87_Semantic_Seg_Mask.zip https://zihao-openmmlab.obs.cn-east-3.myhuaweicloud.com/20230130-mmseg/dataset/watermelon/Watermelon87_Semantic_Seg_Mask.zip

# 使用 Python 解压
import zipfile

with zipfile.ZipFile("Watermelon87_Semantic_Seg_Mask.zip", 'r') as zip_ref:
    zip_ref.extractall(".")

# 删除压缩包
!del Watermelon87_Semantic_Seg_Mask.zip
```
